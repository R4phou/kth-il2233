```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

plt.figure(figsize = (6,3))
n = np.linspace(0, 10, 200)
x = np.sin(math.pi * n)
y = np.sin(math.pi * 0.5 * n)
plt.plot(n, x, c = 'blue')
plt.plot(n, y, c = 'red')
print(np.linalg.norm(x - y))

plt.show()
```
